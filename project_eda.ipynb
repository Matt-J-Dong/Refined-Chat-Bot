```python
import re


# with open("C:\\Users\\thepe\\Downloads\\HP_PhilosophersStone.txt", encoding="utf8") as f:
#     lines = f.readlines()

lines = open("C:\\Users\\thepe\\Downloads\\HP_PhilosophersStone.txt", encoding='utf8').read()

# Split the script into lines
script_lines = lines.strip().split('\n')

# Initialize lists for scene descriptions and dialogue lines
scene_descriptions = []
dialogue_lines = []

# Iterate through the lines and separate scene descriptions and dialogue
for i in range(0, len(script_lines)):
    if script_lines[i].strip() != '':
        # print(script_lines[i])

        if script_lines[i].startswith('['): # this is description
            scene_descriptions.append(script_lines[i].strip('[]'))
        else:
            dialogue_lines.append(script_lines[i])
```

```python
# Print the results
print("Scene Descriptions:")
for scene in scene_descriptions:
    print(scene)

print("\nDialogue Lines:")
for line in dialogue_lines:
    print(line)
```

```python
import re
import matplotlib.pyplot as plt

word_pattern = r'\b\w+\b'

sentence_word_counts = [len(re.findall(word_pattern, scene)) for scene in scene_descriptions]

plt.hist(sentence_word_counts, bins=range(0,130,10), edgecolor='black')
plt.title('Sentence length distribution of scene setting descriptions')
```

```python
# characters who talk the most


# Create a dictionary to store character names and their dialogue counts
character_dialogue_counts = {}

# Iterate through the dialogue lines and count character dialogue
for line in dialogue_lines:
    parts = line.split(":")
    if len(parts) >= 2:
        character_name = parts[0].strip()
        if character_name in character_dialogue_counts:
            character_dialogue_counts[character_name] += 1
        else:
            character_dialogue_counts[character_name] = 1

# Sort characters by the number of dialogue lines
sorted_characters = sorted(character_dialogue_counts.keys(), key=lambda x: character_dialogue_counts[x], reverse=True)[:20]
sorted_counts = [character_dialogue_counts[char] for char in sorted_characters]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_characters, sorted_counts, color='b')
plt.xlabel("Number of Dialogue Lines")
plt.ylabel("Characters")
plt.title("Distribution of Characters with the Most Dialogue")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the bar chart
plt.show()
```

```python
word_pattern = r'\b\w+\b'

setting_words = [re.findall(word_pattern, scene) for scene in scene_descriptions]

# plt.hist(sentence_word_counts, bins=range(0,130,10), edgecolor='black')
# plt.title('Sentence length distribution of scene setting descriptions')


setting_word_freq = {}

for line in setting_words:
    for word in line:
        if word in setting_word_freq:
            setting_word_freq[word] += 1
        else:
            setting_word_freq[word] = 0



# Sort characters by the number of dialogue lines
sorted_words = sorted(setting_word_freq.keys(), key=lambda x: setting_word_freq[x], reverse=True)[:20]
sorted_word_counts = [setting_word_freq[char] for char in sorted_words]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_words, sorted_word_counts, color='b')
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.title("Distribution of Word frequency in Setting descriptions")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the bar chart
plt.show()
```

```python
word_pattern = r'\b\w+\b'

dialogue_words = []
for line in dialogue_lines:
    parts = line.split(":")
    if len(parts) >= 2:
        dialogue = parts[1].strip()
        dialogue_words.append(re.findall(word_pattern, dialogue))
# plt.hist(sentence_word_counts, bins=range(0,130,10), edgecolor='black')
# plt.title('Sentence length distribution of scene setting descriptions')


dialogue_word_freq = {}

for line in dialogue_words:
    for word in line:
        if word in dialogue_word_freq:
            dialogue_word_freq[word] += 1
        else:
            dialogue_word_freq[word] = 0



# Sort characters by the number of dialogue lines
sorted_words = sorted(dialogue_word_freq.keys(), key=lambda x: dialogue_word_freq[x], reverse=True)[:20]
sorted_word_counts = [dialogue_word_freq[char] for char in sorted_words]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_words, sorted_word_counts, color='b')
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.title("Distribution of Word frequency in Dialogue")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the bar chart
plt.show()
```
